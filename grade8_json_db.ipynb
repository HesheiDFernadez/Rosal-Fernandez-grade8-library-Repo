```python
import json
with open("books.json", "r") as file:
 data = json.load(file)
print("JSON file loaded")
ref = db.reference("books")
for book in data["books"]:
 ref.child(str(book["id"])).set(book)
print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()
print(books_data)
```

```python
for book_id, info in books_data.items():
 print(info["title"])
```

```python
for book_id, info in books_data.items():
 print(info["title"],
       info["details"] ["available"])
```

```python
print("Total number of books: ", len(books_data))
total_copies = 0
for book_id, info in books_data.items():
  total_copies += info["details"] ["copies"]
print("Total number of books (counting their copies): ", total_copies)
```
